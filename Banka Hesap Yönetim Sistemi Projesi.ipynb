```python
class Kitap:
    def __init__ (self, baslik, yazar, yayin_evi, yayin_yili):
        self.baslik=baslik
        self.yazar=yazar
        self.yayin_evi=yayin_evi
        self.yayin_yili=yayin_yili
        
    def bilgileri_goster(self):
        print("kitap bilgileri: ")
        print(f"baslık: {self.baslik}\nYazar: {yazar}\Yayın evi: {self.yayin_evi}\Yayın yılı: {yayin_yili}")
        
class Uye:
    def __init__ (self, ad, soyad, uye_numarasi):
        self.ad=ad
        self.soyad=soyad
        self.uye_numarasi=uye_numarasi
        
    def bilgileri_goster(self):
        print("uye bilgileri: ")
        print(f"Ad: {self.ad}\nSoyad: {self.soyad}\nÜye Numarası: {self.uye_numarasi}")
        
class Kutuphane:
    def __init__(self):
        self.kitaplar=[]
        self.uyeler=[]
        self.odunc_alinmis_kitaplar={}
        
    def kitap_ekle(self, kitap):
        self.kitaplar.append(kitap)
        print(f"{kitap.baslik} kütüphanemize başarılı bir şekilde eklendi. ")
        
    def kitap_sil(self, baslik):
        for kitap in self.kitaplar:
            if kitap.baslik==baslik:
                self.kitaplar.remove(kitap)
                print(f"{kitap.baslik} baslıklı kitap başarılı bir şekilde silindi. ")
                return 
        print(f"{kitap.baslik} baslıklı kitap bulunamadı")
         
    def uye_ekle(self, uye):
        self.uyeler.append(uye)
        print(f"{uye.ad} {uye.soyad} kütüphanenize başarılı bir şekilde eklendi. ")
        
    def uye_sil(self, uye_numarasi):
        for uye in self.uyeler:
            if uye.uye_numarasi==uye_numarasi: 
                self.uyeler.remove(uye)
                print(f"{uye.uye_numarasi} numarasına sahip kullanıcı silindi")
                return
        print(f"{uye_numarasi} numaraya sahip üye bulunamadı. ")
        
    def odunc_almak(self, kitap, uye):
        if kitap in self.kitaplar and uye in self.uyeler:
            if kitap not in self.odunc_alinmis_kitaplar:
                self.odunc_alinmis_kitaplar[kitap]=uye
                print(f"{kitap.baslik} adlı kitap {uye.ad} {uye.soyad} kişisi tarafından başarılı bir şekilde ödünç alındı. ")
            else:
                print(f"{kitap.baslik} daha önce başka kişi tarafından ödünç alınmıştır. ")
                
    def geri_vermek(self, kitap):
        if kitap in self.odunc_alinmis_kitaplar:
            uye=self.odunc_alinmis_kitaplar.pop(kitap)
            print(f"{kitap.baslik} adlı kitap {uye.ad} {uye.soyad} kişisi tarafından ödünç alınmış")
        else: 
            print(f"{kitap.baslik} odunc alımıs kitaplar listesinde yok. ")
            
    def kitaplar_goster(self):
        if len (self.kitaplar)==0:
            print("henüz kütüphanenizde kitap bulunmamaktadır. ")
        else: 
            print("kitaplar")
            for kitap in self.kitaplar:
                kitap.bilgileri_goster()
                
    def uyeleri_goster(self):
        if len(self.uyeler)==0:
            print("henüz kütüphanenizde üye bulunmamaktadır. ")
        else: 
            print("Uyeler: ")
            for uye in self.uyeler:
                uye.bilgileri_goster()
                
kutuphane = Kutuphane()

while True: 
    print(20*"*", "Kütüphane Yönetim Sistemi", 20*"*")
    print("""
    1. Kitap Ekle
    2. Kitap Sil 
    3. Üye Ekle
    4. Üye Sil 
    5. Kitap ödünç almak
    6. Kitap geri vermek
    7. Kitapları göster
    8. Üyeleri Goster
    9. Çıkış (q)
    """)
    islem=input("Lütfen yapacağınız işlemin numarasını giriniz (1-9): ")
    if islem == "1":
        baslik = input("Başlık giriniz: ")
        yazar = input("Yazar Adi: ")
        yayin_evi = input("Yayın evi: ")
        yayin_yili=input("Yayıl yılı: ")
        kitap=Kitap(baslik, yazar, yayin_evi, yayin_yili)
        kutuphane.kitap_ekle(kitap)
    elif islem == "2":
        baslik =input("Silmek istediğiniz kitabın başlığını giriniz: ")
        kutuphane.kitap_sil(baslik)
    elif islem == "3":
        ad = input("Adı giriniz: ")
        soyad = input("Soyad giriniz: ")
        uye_numarasi=("Üye numarası giriniz: ")
        uye=Uye(ad, soyad, uye_numarasi)
        kutuphane.uye_ekle(uye)
    elif islem=="4":
        uye_numarasi=input("Silmek istediğiniz uyenin uye numarasını giriniz: ")
        kutuphane.uye_sil(uye_numarasi)
    elif islem=="5":
        baslik=input("kitabın adını giriniz: ")
        uye_numarasi=input("Uye numarası giriniz: ")
        kitap=next((k for k in kutuphane.kitaplar if k.baslik==baslik), None)
        uye = next((u for u in kutuphane.uyeler if u.uye_numarasi==uye_numarasi), None)
        if kitap and uye: 
            kutuphane.odunc_almak(kitap, uye)
        else: 
            print("Yanlış bri kitap veya üye girişi yaptınız. ")
    elif islem=="6":
        baslik=input("Kitabın adını giriniz: ")
        kitap=next((k for k in kutuphane.kitaplar if k.baslik==baslik), None)
        if kitap: 
            kutuphane.geri_vermek(kitap)
        else: 
            print("Yanşış bir kitap girdiniz. ")
    elif islem=="7":
        kutuphane.kitaplar_goster()
    elif islem=="8":
        kutuphane.uyeleri_goster()
    elif islem=="9" or islem=="q" or islem =="Q":
        print("Uygulamadan çıkış yapılıyor... ")
        break
    else: 
        print("Yanlış bir işlem numarası girdiniz. ")









class Musteri: 
    def __init__ (self, musteri_numarasi, ad, soyad):
        self.musteri_numarasi=musteri_numarasi
        self.ad=ad
        self.soyad=soyad
        
    def bilgileri_goster(self):
        print("Musteri bilgileri: ")
        print(f"""
            Musteri Numarası: {self.musteri_numarasi}
            Ad: {self.ad}
            Soyad: {self.soyad}
        """)
        
class Hesap: 
    def __init__ (self, hesap_numarasi, musteri, bakiye=0): 
        self.hesap_numarasi=hesap_numarasi
        self.musteri=musteri
        self.bakiye=bakiye
        self.acik_hesap=True
    
    def para_yatirmak(self, miktar):
        if self.acik_hesap and miktar > 0: 
            self.bakiye+=miktar
            print(f"""
            Yatırdığınız miktar: {miktar} TL
            Şuanki hesap bakiyeniz: {self.bakiye} TL
            """)
        elif not self.acik_hesap: 
            print("Hesap kapalıdır. ")
        else:
            print("Girdiğiniz miktar yanlıştır. Lütfen geçerli bir değer giriniz. ")
    
    def para_cekmek(self, miktar):
        if self.acik_hesap and miktar > 0 and miktar <= self.bakiye:
            self.bakiye-=miktar
            print(f"""
            Cektiğiniz miktar: {miktar} TL
            Şuanki hesap bakiyeniz: {self.bakiye} TL
            """)
        elif not self.acik_hesap:
            print("Hesap kapalıdır. ")
        elif miktar > self.bakiye:
            print("Girmiş olduğunuz bakiyenizden fazladır. ")
        else: 
            print("Girdiğiniz bakiye yanlıştır. Lütfen geçerli bir değer giriniz. ")
            
    def bakiye_sorgulama(self):
        if self.acik_hesap:
            print(f"Şu anki hesap bakiyeniz: {self.bakiye} TL")
        else:
            print("Hesap kapalıdır. ")
            
    def hesap_kapatmak(self):
        if self.acik_hesap:
            self.acik_hesap=False
            print("Hesap başarılı bir şekilde kapatıldı. ")
        else: 
            print("Hesap zaten kapalıdır. ")
            
    def hesap_aktif_etmek(self):
        if not self.acik_hesap:
            self.acik_hesap=True
            print("Hesap başarılı bir şekilde aktif edildi. ")
        else: 
            print("Hesap zaten aktiftir. ")
    
    def bilgileri_goster(self):
        print("Hesap bilgileri: ")
        print(f"""
        Hesap Numarası: {self.hesap_numarasi}
        Hesap Bakiyesi: {self.bakiye}
        """)
        self.musteri_bilgileri_goster()
        
class Banka:
    def __init__ (self):
        self.musteriler=[]
        self.hesaplar=[]
        
    def hesap_olusturmak(self, musteri, ilk_bakiye=0):
        hesap_numarasi=len(self.hesap_numarasi) + 100
        hesap=Hesap(hesap_numarasi, musteri, ilk_bakiye)
        self.hesaplar.append(hesap)
        print(f"{hesap_numarasi} başarılı bir şekilde oluşturuldu. ")
        return hesap
    
    def tum_hesaplar(self):
        print("Tüm hesaplar: ")
        for hesap in self.hesaplar: 
            hesap.bilgileri_goster()
            
    def tum_musteriler(self): 
        print("Tüm müsşteriler: ")
        for musteri in self.musteriler:
            musteri_bilgileri_goster()
            
            
banka = Banka()

while True:
    print(20*"*", Banka hesap yönetimi listesi, 20*"*")
    print("""
    1. Hesap oluşturmakü
    2. Tüm hesaplar listesi
    3. Tüm musteriler listesi
    4. Para yatırmak
    5. Para çekmek
    6. Bakiye sorgulama
    7. Hesap Kapatmak (Deactive)
    8. Hesap aktif etmek (Active)
    9. Çıkış (q-Q)
    """)
    islem=input("Lütfen yapacağınız işlemin numarasını giriniz: ")
    
    if islem=="1":
        musteri_numarasi=input("Musteri numarası: ")
        ad=input("Ad: ")
        soyad=input("Soyad: ")
        musteri=Musteri(musteri_numarasi, ad, soyad)
        banka.musteriler.append(musteri)
        banka.hesap_olusturmak(musteri)
    elif islem =="2":
        banka.tum_hesaplar()
    elif islem =="3":
        banka.tum_musteriler()
    elif islem =="4":
        hesap_numarasi=int(input("Hesap numarası: "))
        hesap = next((hesap for hesap in banka.hesaplar if hesap.hesap_numarasi == hesap_numarasi), None)
        if hesap: 
            miktar = float(input("Yatırmak istediğiniz miktarı giriniz: "))
            hesap.para_yatirmak(miktar)
        else:
            print("Hesap bulunamadı")
    elif islem=="5":
        hesap_numarasi=int(input("Hesap numarası: "))
        hesap = next((hesap for hesap in banka.hesaplar if hesap.hesap_numarasi == hesap_numarasi), None)
        if hesap: 
            miktar = float(input("ÇEKMEK istediğiniz miktarı giriniz: "))
            hesap.para_cekmek(miktar)
        else:
            print("Hesap bulunamadı")
    elif islem =="6":
        hesap_numarasi=int(input("Hesap numarası: "))
        hesap = next((hesap for hesap in banka.hesaplar if hesap.hesap_numarasi == hesap_numarasi), None)
        if hesap: 
            hesap.bakiye_sorgulama()
        else:
            print("Hesap bulunamadı")
    elif islem =="7":
        hesap_numarasi=int(input("Hesap numarası: "))
        hesap = next((hesap for hesap in banka.hesaplar if hesap.hesap_numarasi == hesap_numarasi), None)
        if hesap: 
            hesap.hesap_kapatmak()
        else:
            print("Hesap bulunamadı")
    elif islem =="8":
        hesap_numarasi=int(input("Hesap numarası: "))
        hesap = next((hesap for hesap in banka.hesaplar if hesap.hesap_numarasi == hesap_numarasi), None)
        if hesap: 
            hesap.hesap_aktif_etmek()
        else:
            print("Hesap bulunamadı")
    elif islem =="9":
        print("Uygulamadaan çıkış yapılıyor... ")
        break
    else: 
        print("Yanlış bir işlem numarası girdiniz... ")
```
